```python
import json

# Path to the results file
results_path = "<home>/workspace/data/longmemeval_oracle_short100.json"
# Load the JSON data
with open(results_path, "r") as f:
    data = json.load(f)

# Collect all categories
categories = set()
for item in data:
    if "question_type" in item.keys():
        categories.add(item["question_type"])

print(f"Total number of unique categories: {len(categories)}")
print("Categories:")
for cat in sorted(categories):
    print("-", cat)
    
print(len(data))
```

```python
import json

# Path to the results file
results_path = "<home>/workspace/data/longmemeval_s.json"
# Load the JSON data
with open(results_path, "r") as f:
    data = json.load(f)
    
# Define split indices
splits = [(0, 175), (175, 350), (350, 500)]

# Save each split to a separate file, using the starting index in the filename
for start, end in splits:
    split_data = data[start:end]
    filename = f"longmemeval_s_{start}.json"
    with open(filename, "w") as f_out:
        json.dump(split_data, f_out, indent=2)
    print(f"Split {start}-{end}: {len(split_data)} items saved to {filename}")
```
